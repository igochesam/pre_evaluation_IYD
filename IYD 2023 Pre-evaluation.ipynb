### Employment Status

```python
# Get unique values from the employment status column
df['What is your employment status?'].unique()
```

```python
# Get a count of the unique values
df['What is your employment status?'].value_counts()
```

```python
unique_employment_status = df['What is your employment status?'].value_counts().reset_index()

unique_employment_status
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
plt.pie(unique_employment_status['What is your employment status?'], labels=unique_employment_status['index'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(unique_employment_status))))

# Add a title
plt.title('Employment Status')

# Display the chart
plt.show()








```
